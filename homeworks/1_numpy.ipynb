Даны два одинаковой длины вектора (списка) y₁ и y₂, каждый из которых состоит только из 0 и 1.

Левая часть (все элементы до выбранного индекса) — содержит больше совпадающих нулей, чем единиц.
То есть, среди элементов, где y₁[i] == y₂[i], нулей больше, чем единиц.

Правая часть (все элементы после выбранного индекса) — наоборот, содержит больше совпадающих единиц, чем нулей.

Если какая-то из частей оказывается пустой (например, весь массив слева или справа), то она автоматически считается подходящей.

```python
def find_thresholds(y1, y2, debug=False):
    assert len(y1) == len(y2) and len(y1) > 0
    n = len(y1)
    matching_zeros = [1 if y1[i] == y2[i] == 0 else 0 for i in range(n)]
    matching_ones = [1 if y1[i] == y2[i] == 1 else 0 for i in range(n)]
    res = []
    if sum(matching_ones) / n > 0.5:
        res.append(0)
    for i in range(1, n):
        if debug: # если вам непонятна логика работы этой функции, передайте третьим аргументом True
            print(matching_zeros[:i], matching_ones[i:], i)
        if sum(matching_zeros[:i]) / i > 0.5 and sum(matching_ones[i:]) / (n - i) > 0.5:
            res.append(i)
    if sum(matching_zeros) / n > 0.5:
        res.append(n)
    return res
```
