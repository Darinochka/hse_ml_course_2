```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor

# Генерируем обучающие данные
X_train = np.linspace(0, 5, 10).reshape(-1, 1)
y_train = np.sin(X_train).ravel()  # любая функция для примера

# Создаем и обучаем дерево
tree = DecisionTreeRegressor(max_depth=3)  # небольшой рост, чтобы видно было ступенчатость
tree.fit(X_train, y_train)

# Создаем область для визуализации (в том числе за пределами обучающей выборки)
X_plot = np.linspace(-4, 10, 500).reshape(-1, 1)
y_pred = tree.predict(X_plot)

# Визуализация
plt.figure(figsize=(8, 5))
plt.plot(X_plot, y_pred, label="Предсказание дерева", color='orange')
plt.scatter(X_train, y_train, label="Обучающие точки", color='blue', s=50)
plt.axvline(X_train.min(), color='grey', linestyle='--', label="Границы обучающей выборки")
plt.axvline(X_train.max(), color='grey', linestyle='--')
plt.title("Дерево решений не экстраполирует за пределами обучающей выборки")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
